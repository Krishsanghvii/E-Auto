# Facial Recognition Model Research
 **_`Facial Recognition`_** is a complex task that involves detecting and identifying human faces in images or videos. There are several pre-trained models and libraries available that can be leveraged for facial recognition purposes.


###Here are some popular open-source options:

1. __OpenCV__:
  > OpenCV is a widely used open-source computer vision library that includes face detection and recognition capabilities. It offers pre-trained Haar cascades and deep learning-based models like Single Shot Multibox Detector (SSD) and You Only Look Once (YOLO) that can be used for facial recognition.

2. __Dlib__:
> Dlib is a powerful C++ library that provides facial recognition functionality. It includes a pre-trained face recognition model that can compute a 128-dimensional face embedding, which can be used for face matching and identification.

3. __face_recognition__:
> The 'face_recognition' library is a high-level wrapper around dlib's facial recognition capabilities. It allows for easy face detection, face recognition, and emotion analysis. This library is based on the deep learning-based face recognition model implemented in dlib.

4. __MTCNN (Multi-Task Cascaded Convolutional Networks)__:
> MTCNN is a deep learning-based model specifically designed for face detection. It can detect faces in an image with bounding box coordinates and facial landmarks.

5. __FaceNet__:
> FaceNet is a deep learning model developed by Google that computes a 128-dimensional face embedding for face recognition tasks. It has achieved state-of-the-art results in face verification and identification tasks.


# Face Recognition Demo

```python
import face_recognition
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
image_path1 = '/content/images.jpg'
image1 = face_recognition.load_image_file(image_path1)
face_locations1 = face_recognition.face_locations(image1)
plt.imshow(image1)
ax = plt.gca()

for face_location in face_locations1:
    top, right, bottom, left = face_location
    rect = patches.Rectangle((left, top), (right - left), (bottom - top), linewidth=2, edgecolor='r', facecolor='none')
    ax.add_patch(rect)

plt.show()
```

# MTCNN Demo



```python
# !pip install mtcnn
# !pip install matplotlib
```

```python
from mtcnn.mtcnn import MTCNN
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import cv2
```

```python
detector = MTCNN()
```

```python
image_path = '/content/images.jpg'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
faces = detector.detect_faces(image_rgb)
plt.imshow(image_rgb)
ax = plt.gca()

for face in faces:
    x, y, width, height = face['box']
    rect = patches.Rectangle((x, y), width, height, linewidth=2, edgecolor='r', facecolor='none')
    ax.add_patch(rect)

plt.show()
```

```python
image_path = '/content/images1.jpg'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
faces = detector.detect_faces(image_rgb)
plt.imshow(image_rgb)
ax = plt.gca()

for face in faces:
    x, y, width, height = face['box']
    rect = patches.Rectangle((x, y), width, height, linewidth=2, edgecolor='r', facecolor='none')
    ax.add_patch(rect)

plt.show()
```
